```python
import gensim.downloader as api
import numpy as np

# Load pre-trained Word2Vec model
word2vec_model = api.load("word2vec-google-news-300")

# Function to convert a sentence to a Word2Vec embedding
def sentence_to_word2vec(sentence, model):
    words = sentence.lower().split()
    word_embeddings = [model[word] for word in words if word in model]
    if not word_embeddings:
        return np.zeros(model.vector_size)
    return np.mean(word_embeddings, axis=0)

# Sample sentence
sentence = "This is an example sentence."

# Get the sentence embedding
sentence_embedding = sentence_to_word2vec(sentence, word2vec_model)

print("Sentence embedding:", sentence_embedding)
```

# create embedding for each of the text chunck


```python
chunck_vectors = [sentence_to_word2vec(chunck.page_content,word2vec_model) for chunck in text_chuncks]
print(chunck_vectors)
```

```python
# Prepare vectors and metadata for upsert
vectors_to_upsert = [
    {
        "id": f"chunk_{i}",
        "values": vector.tolist(),
        "metadata": {'Text':text.page_content}
    }
    for i, (vector, text) in enumerate(zip(chunck_vectors, text_chuncks))
]
# Upsert vectors into the index
index.upsert(vectors=vectors_to_upsert)

# Optionally, you can print the upsert response
print("Upsert response:", vectors_to_upsert)
```

```python
testing_text = " The dominant sequence transduction models are based on"
testing_vector = sentence_to_word2vec(testing_text,word2vec_model)
testing_vector = testing_vector.tolist()
```

```python
query_response = index.query(vector=testing_vector,top_k=3,include_values=True,include_metadata=True)
```

```python
query_response
query_response = query_response['matches']
query_response
```

```python
response_text = []
for i in query_response:
  res = i["metadata"]
  response_text.append(res["Text"])
response_text
```

```python
TEMPLATE =("""
{text1}{text2}{text3}
based on the above information ans the following quesiton
{question}
""")
```

```python
prompt = PromptTemplate(
    input_variables=["text1","text2","text3","question"],
    template = TEMPLATE
)
from langchain.chains import LLMChain
```

```python
chain = LLMChain(llm=llm,prompt=prompt)
chain.run({"text1":response_text[0],"text2":response_text[1],"text3":response_text[2],"question":testing_text})
```
